### Problem Statement 1:Smart Parking Lot Management System
### Design a function to manage a smart parking lot.
### The system should:
- Accept vehicle entry and exit logs
- Calculate total parked vehicles
- Identify peak parking usage
- Alert if parking exceeds capacity

```python
def check_parking(req,logs):
    in_veh=0  # current vehicles
    peak=0  # peak vehicles
    for log in logs:
        if(log.lower()=='in'):
            in_veh+=1
            peak+=1
        elif(log.lower()=='out'):
            in_veh-=1
    print("Peak ussage of parking is: ",peak)
    if(in_veh>=req):  # capacity check
        print("Alert parking capacity exceeded")
    else:
        print("Current parked vehicles ",in_veh)
        print("Parking : Available")
req=int(input("Enter the parking lot capacity: "))
logs=list(map(str,input().split()))
check_parking(req,logs)
```

### Problem Statement 2:Online Food Delivery Time Estimator
### Create a function that estimates delivery time based on:
- Distance (km)
- Weather condition
- Traffic level
- Apply delays dynamically and display final ETA.
- Real-Time Use
- Food delivery apps

- Logistics platforms

```python
def timeestimator(dist,traffic,weather):
    time=5  # base time per km
    time=time*dist  # initial time based on distance
    if(traffic.lower()=='high'):
        time=time*2  # high traffic multiplier
    elif(traffic.lower()=='medium'):
        time=time*1.5  # medium traffic multiplier
    if(weather.lower()=='rainy'):
        time=time*2  # rainy weather multiplier
    elif(weather.lower()=='sunny'):
        time=time*1.5  # sunny weather multiplier
    elif(weather.lower()=='normal'):
        time=time*1
    return time
dist=int(input("Enter the distance in km: "))
traffic=(input("Enter the traffic amount like low ,medium,high: "))
weather=input("Enter the weather Rainy/Sunny/Normal: ")
print("Estimated Time for Delivery: ",timeestimator(dist,traffic,weather))
```

### Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
### Build a function that analyzes seat booking data and:
- Calculates occupancy percentage
- Determines if show is Housefull
- Suggests opening additional shows

```python
def costcal(seats,occupancy):
    total=0  # count occupied seats
    for occupied in occupancy:
        if(occupied):
            total+=1
    per=(total*100)/seats  # calculate occupancy percentage
    print("Occupancy percentage is: ",per)
    if(per==100):
        print("HouseFull ,open additional shows")
    elif(per>=75 and per<=99):
        print("Almost Full")
    else:
        print("Seats available")
seats=int(input("Enter the total no of seats: "))
occupancy=list(map(int,input("Enter the occupancy 1 for occupied and 0 for not occupied separated by a space").split()))
costcal(seats,occupancy)
```

###  Problem Statement 4:Cloud Server Load Classification System
### Create a function to classify server load based on CPU usage readings.
### Rules:
- Average CPU < 50% → Normal
- 50-–80% → Warning
- 80% → Critical

```python
def calcload(cpu_readings):
    sum1=sum(cpu_readings)  # total CPU readings
    avg=(sum1)/len(cpu_readings)  # calculate average load
    print("CPU load is: ",avg,"%")
    if(avg<50):
        print("Server Status: NORMAL LOAD")
    elif(avg>=50 and avg<=80):
        print("Server Status: WARNING")
    else:
        print("Server Status: CRITICAL")
cpu_readings=list(map(int,input("Enter the readings seperated by a space: ").split()))
calcload(cpu_readings)
```

### Problem Statement 5:Smart Classroom Resource Usage Monitor
### Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.
### Real-Time Use
- Smart classrooms
- Energy optimization systems

```python
def findoveruse(Resource_Usage):
    item=None  # store overused resource
    val=0  # store highest usage
    for resource in Resource_Usage:
        if(val<Resource_Usage[resource]):
            val=Resource_Usage[resource]
            item=resource
    print("Overused Resource is: ",item)
    if(val>6 and val<=12): #assuming classroom is of 12hrs
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")
Resource_Usage={
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}
findoveruse(Resource_Usage)
```

###  Problem Statement 6: Online Event Registration Capacity Controller
### Create a function that manages event registrations by:
- Tracking registrations
- Preventing overbooking
- Triggering waitlist mode

```python
def eventmanage(eventcap,reg):
    if(reg< eventcap):  # registrations below capacity
        print("Confirmed Registerations: ",reg)
        print("Waitlist: 0")
        print("Registeration Status: Open")
    elif(reg==eventcap):  # registrations equal to capacity
        print("Confirmed Registerations: ",reg)
        print("Waitlist: 0")
        print("Registeration Status: Closed")
    else:  # registrations exceed capacity
        print("Confirmed Registerations: ",eventcap)
        print("Waitlist: ",reg-eventcap)
        print("Registeration Status: Closed")
        
eventcap=int(input("Enter the event capacity: "))
reg=int(input("Enter the total no of registerations"))
eventmanage(eventcap,reg)
```
