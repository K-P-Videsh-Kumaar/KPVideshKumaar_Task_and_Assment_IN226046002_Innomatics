### 2. Pass / Fail Analyzer
### Problem Statement:
### Analyze student results.
### marks = [45, 78, 90, 33, 60]
### Requirements:
### A student passes if marks â‰¥ 50
### Count the total number of pass students
### Count the total number of fail students
### Print the final result clearly


```python
marks = list(map(int, input("Enter the marks" ).split())) #taking input
pass_no=fail_no=0 #initialising pass and fail to 0
for i in marks:
    if(i>=50):
        pass_no+=1
    else:
        fail_no+=1
print("The total no of pass students are ",pass_no)
print("The total no of fail students are ",fail_no)
```

### 3. Simple Data Cleaner
### Problem Statement:
### Clean and standardize user names.
### names = [" Alice ", "bob", " CHARLIE "]

### Requirements:
### Remove extra spaces
### Convert all names to lowercase
### Print the cleaned list

```python
n=int(input("Enter the no of names")) #taing no of entries
names=[] #initalising list
for i in range(0,n):
    x=input()
    names.append(x) #inserting into the list
for i in range(0,len(names)):
    names[i]=names[i].strip().lower() # strip to remove spaces and lower to convert into lower case
print(names)
```

### 4. Message Length Analyzer
### Problem Statement:
### Analyze message sizes.
### messages = ["Hi", "Welcome to the platform", "OK"]

### Requirements:
### Print the length of each message
### Flag messages longer than 10 characters

```python
messages=list(map(str,input("Enter the message ").split(',')))
for message in messages:
    print(message, " Length-",len(message))
    if(len(message)>10):
        print(message," is flagged for crossing 10 charecters.")
```

### 5. Error Message Detector
### Problem Statement:
### Detect error messages from system logs.
### logs = ["INFO", "ERROR", "WARNING", "ERROR"]

### Requirements:
### Count the number of "ERROR" entries
### Print the total error count

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for log in logs:
    if log.upper()=="ERROR":
        count+=1
print("The Total Error Count is ",count)
```
