#### Problem 1: Unique Words in a Sentence
#### Problem Statement
- Given a sentence, identify all unique words and display:
- The count of unique words
- The unique words themselves

```python
message="python is easy and python is powerful"
l1=message.split() #split function splits the words based on delimiter and adds it to a list
unique=set() #set function creates a empty set
for i in l1:
    unique.add(i)
print("The total no of unique words are",len(unique)) # calculate length of the set
print (unique)
```

### Problem 2: Highest Salary from Employee Data
- Problem Statement
- Store employee names and their salaries, and determine which employee has the highest salary.


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
name=""
val=0
for employee in employees:
    if employees[employee]>val: #Check if the salary greater than value
        name=employee
        val=employees[employee]
print("Highest Salary : ",name,val)
```

### Problem 3: Find Maximum and Minimum Values
- Problem Statement
- Find the maximum and minimum values from a list of numbers.

```python
numbers = [45, 22, 89, 10, 66]
max1=numbers[0]
min1=numbers[0]
for num in numbers:
    max1=max(max1,num)
    min1=min(min1,num)
print("Highest Number is ",max1)
print("Lowest Number is ",min1)
```

### Problem 4: Count Products Above a Price Threshold
- Problem Statement
- Count how many products have a price greater than 1000.


```python
prices = [450, 1200, 899, 1500, 300]
count=0
for price in prices:
    if price>1000:
        count+=1
print(count," products have a price greater than 1000")
```

### Problem 5: Calculate Attendance Percentage
- Problem Statement
- Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attendance = ["P", "P", "A", "P", "P"]
count=0
for i in attendance:
    if i=="P":
        count+=1
print("Attendence percentage is ",(count/len(attendance)*100))
```

### Problem 6: Remove Duplicate Phone Numbers
- Problem Statement
- You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display   only the unique phone numbers.

```python
phone_numbers = [9876543210, 9123456789, 9876543210]
removed_duplicates=set(phone_numbers)
print("Unique phone numbers are: ",removed_duplicates)
```

### Problem 7: Count Character Frequency
- Problem Statement
- Count how many times each character appears in a given string.


```python
string =input("Enter the string: ")
d={}
for char in string:
    d[char]=d.get(char,0)+1 #get function is used to get the value from the dict and if it is not there replaces it with 0
print(d)
```

### Problem 10: Calculate Average Marks
- Problem Statement
- Calculate the average marks of a student from a list of marks.

```python
marks=list(map(int,input("Enter the marks: ").split()))
sum1=0
avg=0
for mark in marks:
    sum1+=mark
avg=sum1/len(marks)
print("Average marks are: ",avg)
```
