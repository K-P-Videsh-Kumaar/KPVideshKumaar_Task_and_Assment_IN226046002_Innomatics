### Problem 3: Agriculture – Rainfall Adequacy Checker
### Problem Statement
### Create a function that analyzes seasonal rainfall data.
### Rules
- Calculate average rainfall
- Average ≥ required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall

```python
def rainfallcheck(rains,req):
    # Calculate total rainfall
    sum1=0
    for rain in rains:
        sum1+=rain
    print("Average Rainfall ",sum1/len(rains))  # Print average rainfall
    # Check if average meets required rainfall
    if(sum1/len(rains) >= req):
        print("Adequate Rainfall")
    else:
        print("Inadequate Rainfall")

rainfall_data = [100, 95, 105, 100, 100]  # Sample rainfall data
req=100  # Required average rainfall
rainfallcheck(rainfall_data,req)  # Run the check
```

### Problem 4: Social Media – Duplicate Account Detection
### Problem Statement
### Create a function to detect duplicate usernames in a system.
### Rules
- If duplicates exist → Flag accounts
- Else → Accounts are unique

```python
def socialdupli(accounts):
    # Check for duplicate accounts
    flag=False
    s=set()
    for account in accounts:
        if account in s:
            flag=True
            print("Duplicate Account : ", account)  # Print duplicate account
        else:
            s.add(account)
    # Print summary of duplicates
    if(flag):
        print("All Duplicate Usernames flagged")
    else:
        print("Accounts are unique")
# Test with sample account list
socialdupli(["john123", "emma_w", "alex99", "john123", "mike07", "emma_w"])
```

### Problem 6: Agriculture – Premium Crop Price Filter
### Problem Statement
### Create a function that filters crop prices above a premium threshold.
### Rules
- Price > ₹2000 → Premium Crop
- Else → Regular Crop

```python
def pricecheck(crops):
    # Separate crops into premium and regular based on price
    premium=[]
    regular=[]
    for crop in crops:
        if crop>2000:
            premium.append(crop)  # Add to premium if price > 2000
        else:
            regular.append(crop)  # Add to regular otherwise
    print("Premium Crops are: ",premium)
    print("Regular Crops are: ",regular)
# Sample crop prices
crops = [1500, 2200, 1800, 2500, 3000, 1900]
pricecheck(crops)
```

### Problem 7: System Monitoring – Application Health Checker
### Problem Statement
### Rules
- Errors = 0 → Healthy
- Errors ≤ 5 → Minor Issues
- Errors > 5 → Critical Issues

```python
def errorcheck(error):
    # Return system status based on error value
    if(error==0):
        return "Healthy"  # No errors
    elif(error<=5):
        return "Minor Issues"  # Small number of errors
    else:
        return "Critical Issues"  # Many errors
# Test with different error values
print("Error: 0",errorcheck(0))
print("Error: 5",errorcheck(5))
print("Error: 7",errorcheck(7))
```

- Daily limit = ₹50,000

### Problem 9: E-Learning – Student Attendance Eligibility System
### Problem Statement
### Create a function to calculate attendance percentage and check exam eligibility.
### Rules
- Attendance ≥ 75% → Eligible
- Attendance < 75% → Not Eligible

```python
def attendencecheck(attendence):
    # Calculate attendance percentage
    count=0
    for att in attendence:
        if(att.upper()=='P'):
            count+=1  # Count present days
    return count/len(attendence) *100  # Return percentage
# Sample attendance list
attendance = ['p', 'a', 'P', 'A', 'p']
print("Attendence percentage is: ", attendencecheck(attendance))   # Check exam eligibility based on attendance
if(attendencecheck(attendance)>=75):
    print("Exam Eligibility: Eligible")
else:
    print("Exam Eligibility: Not Eligible")
```

### Problem 10: Smart Electricity Bill Analyzer
### Problem Statement
### You are building a smart electricity billing system for residential users.
### Create a function that calculates the electricity bill and classifies usage level based on units consumed.
### Billing Rules
- First 100 units → ₹3 per unit
- Next 100 units (101–200) → ₹5 per unit
- Above 200 units → ₹7 per unit
### Usage Classification
- Total bill < ₹500 → Low Usage
- ₹500 – ₹1500 → Moderate Usage
- ₹1500 → High Usage

```python
def billcalc(units):
    # Calculate electricity bill based on units consumed
    if(units<=100):
        return units*3  # Rate 3 per unit for first 100 units
    if(units>100 and units<=200):
        return (100*3+ (units-100)*5)  # Rate 5 per unit for next 100 units
    if(units>200):
        return (100*3 + (100)*5+ (units-200)*7)  # Rate 7 per unit beyond 200 units
# Sample units consumed
units=150
bill=billcalc(units)
print("Bill Amount is: ",bill)
# Categorize usage based on bill amount
if(bill<500):
    print("Low Ussage ")
elif(bill >=500 and bill <1500):
    print("Moderate Ussage ")
else:
    print("High Ussage ")
```
