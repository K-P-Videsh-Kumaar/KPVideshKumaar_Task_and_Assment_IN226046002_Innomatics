## Task 2:Build Advanced Preprocessing Function

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def preprocess_text(text):
    if not text or text.strip()=="":
        return []
    text=re.sub(r'\d+', '', text)
    text=text.rstrip()
    text=text.lstrip()
    text=text.lower()
    text=re.sub(r'[^\w\s]','',text)
    text=text.translate(str.maketrans('','',string.punctuation))
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    stop_words=set(stopwords.words('english'))
    stop_words = stop_words - {"not", "no", "never"}
    tokenisation=text.split()
    tokens=[word for word in tokenisation if word not in stop_words]
    token = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    return token
```

## Task 3: Stress Testing

```python
lst=[
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
tokens=[]
sentences=[]
for i in lst:
    print(f"\n Original Text:- {i}")
    res=preprocess_text(i)
    tokens.extend(res)
    print(f"Cleaned Tokens:- {res}")
    print(" ".join(res))
    sentences.append(" ".join(res))
```

## Task 4: Token Analytics

```python
for i in lst:
    res=preprocess_text(i)
    print(f"Total no of tokens:- {len(res)}")
    print(f"Number of unique tokens:- {len(set(res))}")
    avg_len = sum(len(t) for t in res) / len(res) if len(res) > 0 else 0
    print(f"Average token length:- {avg_len}")
```
