### Problem 1: Mobile Recharge Validation System
###  Problem Statement
### Create a function to validate a mobile recharge request.
### Rules:
- Recharge amount must be ≥ ₹50
- Valid plans: 199, 299, 399, 599
- User should be allowed to retry if invalid

```python
# Function to check if amount is a valid plan
def isvalid(n):
    valid=[199,299,399,599] # List of valid recharge plans
    if(n in valid):
        return True
    return False

flag=False #for loop to run until successfull recharge
while(not flag):
    amount=int(input("Enter the amount : ")) # Take user input
    if(amount<50):  # Minimum amount check
        print("Enter the amount greater than 50")
    elif isvalid(amount): 
        flag=True
        print("Recharge successfull for ",amount) # Success message
    else:
        print("Plan not available")  # Invalid plan message
```

### Problem 2: Inventory Reorder Alert System
### Problem Statement
### Given a dictionary of products and stock quantities:
- Print Reorder Alert if stock < 15
- Print Stock OK otherwise

```python
# Function to check if stock is below minimum level
def check_stock(stock):
    if(stock<15):
        return True
    return False
    
# Dictionary storing product names and their stock
products={"Milk":20 ,"Egg Trays":13,"Rice Bags":30,"Wheat Bags":2}

# Loop through each product
for product in products:
    if(check_stock(products[product])):
        print("Reorder alert ",product, " stock is less than 15")  # Low stock message
    else:
        print(product," Stock OK ")  # Sufficient stock message
```

### Problem 3: Student Result Processing System
### Problem Statement
### Create a function that:
- Accepts a list of student marks
- Calculates average marks
- Prints Pass if average ≥ 50 else Fail

```python
# Function to calculate average marks
def calculate(marks):
    avg=0
    for mark in marks:
        avg+=mark   # Add each mark
    return avg/len(marks)   # Return average

# Take marks input from user
marks=list(map(int,input("Enter the marks seperated by space: ").split()))

# Check pass or fail based on average
if(calculate(marks)>=50):
    print("Pass")
else:
    print("Fail")
```

### Problem 4: Cab Fare Estimator with Retry Option
### Problem Statement
### Create a cab fare calculator:
- Base fare = ₹50
- ₹12 per km
- Peak hour → 25% extra
- Allow retry using while loop

```python
# Function to calculate total fare
def calculate_fare(kilometer, hour):
    hours=[5,6,7,8,9]   # Peak hours
    fare=50+12*kilometer   # Base fare + per km charge 
    if(hour in hours):
        peak=fare*0.25   # 25% peak charge
        return fare+peak
    return fare   # Normal fare

retry=True   # Control variable for loop

# Loop to calculate fare repeatedly
while(retry):
    km=int(input("Enter the kilometers: "))   # Input distance
    hr=int(input("Enter the time only hrs: "))   # Input hour    
    print("Total Fare is ",calculate_fare(km,hr))   # Display fare
    print("To Stop press -1 or else press 1")
    ch=int(input())
    if(ch==-1):
        retry=False   # Stop the loop
```

### Problem 5: Employee Attendance Eligibility Checker
### Problem Statement
### Create a function that:
- Accepts attendance list ("P", "A")
- Calculates attendance percentage
- Returns Eligible if ≥ 75% else Not Eligible

```python
# Function to calculate attendance percentage
def check_attendance(att):
    count=0
    for value in att:
        if value=="P":
            count+=1   # Count present days
    return count/len(att)*100   # Return attendance percentage

# Take attendance input (P for Present, A for Absent)
attlist= list(map(str,input("Enter the Attendece P or A: ").split()))

# Check eligibility based on 75% criteria
if(check_attendance(attlist)>=75):
    print("Eligible")
else:
    print("Not Eligible")
```

### Problem 6: Password Strength Checker
### Problem Statement
### Create a function to check password strength.
### Rules:
- Minimum length = 8
- Must contain at least one digit
- Must contain at least one special character (@#$)

```python
# Function to check password strength
def checkpass(password):
    if(len(password)>=8):   # Check minimum length
        digit=False
        spchar=False
        for char in password:
            if(char >='0' and char<='9'):
                digit=True   # Check for digit
            elif(char=='@' or char=='#' or char=='$'):
                spchar=True   # Check for special character
        if(digit and spchar):
            return True   # Strong password
    return False   # Weak password

password=input()   # Take password input

# Display result
if(checkpass(password)):
    print(password, " is a strong password ")
else:
    print(password, "is not a strong password ")
```
