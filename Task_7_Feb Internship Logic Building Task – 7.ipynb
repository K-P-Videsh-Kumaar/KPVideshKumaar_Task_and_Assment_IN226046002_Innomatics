### Problem 1: Smart Light Controller
### Problem Statement
- Create a class to control a smart light that can be turned ON or OFF and display its self.status.

```python
class Light: ## class keyword is used to define a class
    def __init__(self,status): ## _init_ is the constructor for initializing class members
        self.status=status ## self keyword is used to access members of class
    def checklight(self):
        if(self.status.lower()=='off'):
            print("status is off,You can turn it on")
        elif(self.status.lower()=='on'):
            print("status is on, You can turn it off")
action=input("Enter the action: ")
var=Light(action) ## To create object of a class in main function we use ClassName(parameter1, parameter2, ...)
var.checklight()
```

### Problem 2: Employee ID Card System
### Problem Statement
- Create a class that stores employee name, ID, and department and displays ID card details.

```python
class employee:
    def __init__(self,name,ID,department):
        self.name=name
        self.ID=ID
        self.department=department
    def display(self):
        print("Employee ID Card:")
        print("Employee name: ",self.name)
        print("Employee ID: ",self.ID)
        print("Department: ",self.department)
name=input("Enter the name: ")
ID=input("Enter the ID: ")
department=input("Enter the Department: ")
emp=employee(name,ID,department)
emp.display()
```

### Problem 3: Mobile Contact Record
### Problem Statement
- Create a class to store contact name and phone number and display contact information.

```python
class mobilerecord:
    def __init__(self,name,number):
        self.name=name
        self.number=number
    def display(self):
        print("Contact Saved")
        print("Name: ",self.name)
        print("Phone Number: ",self.number)
name=input("Enter your name: ")
number=int(input("Enter the number"))
mob=mobilerecord(name,number)
mob.display()
```

### Problem 4: Product Price Tag Generator
### Problem Statement
- Create a class that stores product name and price and prints a formatted price tag.

```python
class product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def display(self):
        print("Product: ",self.name)
        print("Price: ₹",self.price)
name=input("Enter product name: ")
price=input("Enter price of the product: ")
pro=product(name,price)
pro.display()
```
